```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

SQUARE_SIZE = 10
COUNT = 0
TOTAL = 0
an = np.linspace(0, np.pi/2, 100)

def generate_point(size):
    x = random.random()*size
    y = random.random()*size
    return (x, y)

def in_circle(point, size):
    return math.sqrt(point[0]**2 + point[1]**2) <= size

def calculate_pi(total, count):
    return (count / total) * 4
```

```python
SAMPLE = 1000

plt.axes().set_aspect('equal')
plt.plot(10*np.cos(an), 10*np.sin(an))

for i in range(SAMPLE):
    point = generate_point(SQUARE_SIZE)
    plt.plot(point[0], point[1], 'r.')
    TOTAL += 1
    if in_circle(point, SQUARE_SIZE):
        COUNT += 1

print(calculate_pi(TOTAL, COUNT))
print('Total: %d, Inside: %d, Outside: %d' % (TOTAL, COUNT, TOTAL-COUNT))
```
